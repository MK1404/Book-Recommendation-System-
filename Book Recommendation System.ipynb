```python
ratings= pd.read_csv('Ratings.csv')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'book_rating'},inplace=True)
```

```python
users=pd.read_csv('Users.csv')
users.head(2)
```

```python
users.rename(columns={'User_ID':'user_id','Location':'location','Age':'age'},inplace=True)
```

```python
users.head(2)
```

```python
ratings.rename(columns={'book_rating':'rating'},inplace=True)
```

```python
ratings.head(2)
```

```python
print("shape of Books Dataframe :",books.shape)
print("shape of users Dataframe :",users.shape)
print("shape of Ratings Dataframe :",ratings.shape)
```

```python
# find ratings according to users counts
x=ratings['user_id'].value_counts()>200
```

```python
y = x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'Number of Ratings'},inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['Number of Ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric = 'cosine',algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
def recommender(book_name):
    book_id=np.where(book_pivot.index == book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=7)
    for i in range(len(distances.flatten())):
        if i == 0:
              print('Recommendations for {0}:\n'.format(book_pivot.index[book_id]))
        else:
            print('{0}: {1}, with distance of {2}:'.format(i, book_pivot.index[suggestions.flatten()[i]], distances.flatten()[i])) 

            
```

```python
recommender('Harry Potter and the Chamber of Secrets (Book 2)')
```
